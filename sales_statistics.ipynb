### Import Libraries for analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Statistics on sales data
1. Import data 
2. min, max, median and std
3. percentiles and IQR
4. histogram of sales price
5. histogram of IQR sales
6. simulation of random dataset based on mean and std
7. histogram of random sales data

```python
full_data = pd.read_csv('..//Data//full_dataset.csv')
```

```python
sales_list = np.array(full_data.price)
sales_list_sorted = np.sort(sales_list)
```

```python
sales_mean = round(sales_list_sorted.mean(),2)
sales_med = np.median(sales_list_sorted)
sales_min = min(sales_list_sorted)
sales_max = max(sales_list_sorted)
sales_std = round(sales_list_sorted.std(),2)
```

```python
print ('The number of orders is '+str(len(sales_list_sorted)))
print ('The minimum order is '+str(sales_min))
print ('The median order is '+str(sales_med))
print ('The average order is '+str(sales_mean))
print ('The maximum order is '+str(sales_max))
print ('The standard deviation is '+str(sales_std))
print ('The 25th percent quartile is '+str(np.percentile(sales_list, 25)))
print ('The 50th percent quartile is '+str(np.percentile(sales_list, 50)))
print ('The 75th percent quartile is '+str(np.percentile(sales_list, 75)))
```

```python
plt.figure(figsize=(8,5))
plt.hist(sales_list_sorted, bins=20, color='black',edgecolor='k', alpha=0.30)
plt.axvline(sales_mean, color='red', linestyle='dashed', linewidth=1, label = ('Mean: '+str(sales_mean)))
plt.axvline(sales_med, color='blue', linestyle='dashed', linewidth=1, label = ('Median: '+str(sales_med)))
plt.xlabel('Sales Price')
plt.ylabel('No. of Orders')
plt.title('Etsy Sales Data Analysis')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(sales_list_sorted, bins=10, range=(10,29),color='black',edgecolor='k', alpha=0.30)
plt.axvline(sales_mean, color='red', linestyle='dashed', linewidth=1, label = 'Mean = 23.48')
plt.axvline(sales_med, color='blue', linestyle='dashed', linewidth=1, label = 'Median = 13.00')
plt.xlabel('Sales Price')
plt.ylabel('No. of Orders')
plt.title('IQR inclusive Sales Data')
plt.legend()
plt.show()
```

```python
sim = np.random.normal(sales_mean, sales_std, size=10000)
sim.mean()
```

```python
plt.figure(figsize=(8,5))
plt.hist(sim, range=(0,125),color='red',edgecolor='k', alpha=0.30)
plt.xlabel('Sales Price')
plt.ylabel('No. of Orders')
plt.title('Simulated Sales Data, n=10k')
plt.show()
```
