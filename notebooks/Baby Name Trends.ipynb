## Importing libraries to conduct the EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Top 100 baby names by year and gender
Data sourced from https://data.gov.au/

```python
data = pd.read_csv('20180718_bdm_top-100-baby-names-1960-to-2005.CSV')
```

### Exploring the data

```python
data.head()
#data.tail()
##data.shape
#data.describe
#data.dtypes
#data.Year.min()
#data.Year.max()
#data.Year.value_counts()
```

### Cleaning the data

```python
data.columns = map(str.lower, data.columns)
```

```python
data = data.rename(columns={'sex':'gender'})
```

### So what's in a name?

```python
data[data.name=='BENJAMIN'].sort_values(by='year')
```

### So what were the most popular boys names in 1982

```python
data[(data.year == 1982) & (data.gender=='Male')].sort_values(by=['count'],ascending=False).head(10)
```

### Lets visualise this!

```python
#data[data.name=='BENJAMIN'].plot(kind='bar',x='year',figsize=(10,4))
data[data.name=='BENJAMIN'].sort_values(by=['year'],ascending=True).plot(kind='bar',x='year',figsize=(10,4))
plt.title('Show me those Benjamins')
plt.show()
```

### Look at popular current names (well 2005 current...)

```python
data[(data.year == 2005) & (data.gender=='Female')].sort_values(by=['count'],ascending=False).head(10)
```

```python
emily = data[data.name=='EMILY'].sort_values(by=['year'],ascending=True)
```

```python
emily.head()
```

```python
emily.plot(kind='bar',x='year',figsize=(10,4))
plt.title('Trending')
plt.show()
```
