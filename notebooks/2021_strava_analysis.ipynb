### 2021 Strava Running Analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('..//Data//strava_run_data.csv')
```

```python
data = data[data['year']==2021]
```

```python
data.run_category.value_counts()
```

```python
plt.hist(data=data, x='distance_km',bins=40, color='orange',alpha=0.60,edgecolor='k')
plt.title('Distribution of Distance per run')
plt.show()
```

```python
plt.hist(data=data, x='distance_km',bins=10, color='orange',alpha=0.60,edgecolor='k',range=(0,20))
plt.title('Distribution of Distance per run (0-20km)')
plt.show()
```

```python
plt.hist(data=data, x='distance_km',bins=20, color='orange',alpha=0.60,edgecolor='k',range=(20,90))
plt.title('Distribution of Distance per run (20km+)')
plt.show()
```

```python
inc_mean_dist = data.distance_km.expanding().mean()
```

```python
plt.plot(inc_mean_dist)
plt.title('Incremental Run Distance average')
plt.show
```

```python
monthly_distance = data.groupby('month').agg({'distance_km':'sum'})
```

```python
monthly_distance.plot.bar(color='orange',alpha=0.60,edgecolor='k')
plt.title('Monthly Distance')
plt.show
```

```python
monthly_distance
```

```python
monthly_vert = data.groupby('month').agg({'total_elevation_gain':'sum'})
```

```python
monthly_vert.plot.bar()
plt.title('Monthly Vertical Gain')
plt.show
```

```python
x = np.array(data.distance_km)
y = np.array(data.total_elevation_gain)

plt.scatter(x, y, color='Orange',alpha=0.60)
plt.title('Distance & Vertical')
plt.xlabel('Distance')
plt.ylabel('Vertical Gain')
plt.show()
```
